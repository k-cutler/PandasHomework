```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#find total number of unique players using len of "SN"
player_count = len(purchase_data["SN"].unique())

#create a 'summary' table with column "Total Players"
total_players = pd.DataFrame({"Total Players":[player_count]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find total number of purchases based on Purchase ID
total_purchases = len(purchase_data["Purchase ID"].unique())

#find number of unique items based on Item ID
unique_items = len(purchase_data["Item ID"].unique())

#find total revenue using based on sum of Price
total_rev = purchase_data["Price"].sum()

#find average price based on mean of Price
avg_price = purchase_data["Price"].mean()

#create summary table with new variables
summary_purchases = pd.DataFrame({"Number of Unique Items":[unique_items],
                                 "Average Price":[avg_price],
                                 "Number of Purchases":[total_purchases],
                                 "Total Revenue":[total_rev]})
#clean avg price and total revenue
summary_purchases["Average Price"] = summary_purchases["Average Price"].map("${:,.2f}".format)
summary_purchases["Total Revenue"] = summary_purchases["Total Revenue"].map("${:,.2f}".format)
summary_purchases
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#need to remove duplicate user purchases
#https://www.geeksforgeeks.org/python-pandas-dataframe-drop_duplicates/
no_duplicates = purchase_data.drop_duplicates(subset="SN",keep='first',inplace=False)

#using value_counts from the no duplicates variable, find the count of each gender. Store this as a data frame
gender_count = pd.DataFrame(no_duplicates["Gender"].value_counts())

#calculate the percentage of each gender from gender count divided by the player count from above
gender_pct = round(gender_count/player_count*100,2)

#using the percentages, add a column for Percentage of players
gender_count["Percentage of Players"] = gender_pct

#rename Gender to Total Count
gender_dems = gender_count.rename(columns={"Gender":"Total Count"})
gender_dems
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create gender group by object from Purchase Data DF column "Gender"
gender_group = purchase_data.groupby(["Gender"])

#calculate purchase count by gender
purchase_count = gender_group["SN"].count()

#calculate average price paid for items by each gender
avg_price_gender = round(gender_group["Price"].mean(),2)

#calculate the total purchase value by gender
total_paid_gender = gender_group["Price"].sum()

#create group by object from no_duplicates by "Gender" so we can calculate avg total purchase per person
gender_no_duplicates = no_duplicates.groupby(["Gender"])

#find avg total purchase per person from total_paid_gender divided by count of "SN" in gender_no_duplicates
avg_total_purchase = round(total_paid_gender/gender_no_duplicates["SN"].count(),2)

#create data frame
summary_df = pd.DataFrame(gender_group["SN"].count())

#add columns to data frame
summary_df["Average Purchase Price"] = avg_price_gender.map("${:,.2f}".format)
summary_df["Total Purchase Value"] = total_paid_gender.map("${:,.2f}".format)
summary_df["Avg Total Purchase per Person"] = avg_total_purchase.map("${:,.2f}".format)

#rename SN column to Purchase Count
purchase_analysis_gender_df = summary_df.rename(columns={"SN":"Purchase Count"})
purchase_analysis_gender_df
```

* Create a summary data frame to hold the results

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create bins based on age
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]

#create labels for each bin
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#store purchase_data into new variable so we can work with bins
age_purch_analysis_df = purchase_data

#create column with bins
age_purch_analysis_df["Age Demographics"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

#create groupby object on Age Demographics
age_purch_analysis_df = age_purch_analysis_df.groupby("Age Demographics")

#create summary
summary_age_purch_analysis = pd.DataFrame(age_purch_analysis_df["Purchase ID"].count())
summary_age_purch_analysis

#calculate total purchase value by age
total_paid_age = age_purch_analysis_df["Price"].sum()

#calculate averge purchase price by age
avg_purch_age = round(age_purch_analysis_df["Price"].mean(),2)

#calculate avg total purchase per person by age from total paid by age divided by no duplicates purchase 
#count by age, similar to what we did for gender
avg_total_purch_age = round(total_paid_age/age_demos_df["Purchase ID"].count(),2)

#summary dataframe
summary_age_purch_2 = pd.DataFrame(age_purch_analysis_df["SN"].count())

#add columns for Average Purhcase Price, Total Purchas Value, Avg Total Purchase per Person
summary_age_purch_2["Average Purchase Price"] = avg_purch_age.map("${:,.2f}".format)
summary_age_purch_2["Total Purchase Value"] = total_paid_age.map("${:,.2f}".format)
summary_age_purch_2["Avg Total Purchase per Person"] = avg_total_purch_age.map("${:,.2f}".format)

#final summary table renaming SN to Purchase Count
age_purch_analysis_df_2 = summary_age_purch_2.rename(columns={"SN":"Purchase Count"})
age_purch_analysis_df_2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create group by object from "SN" in purchase_data DataFrame
sn_grouped = purchase_data.groupby("SN")

#create a DataFrame with grouping
sn_grouped_df = pd.DataFrame(sn_grouped["Purchase ID"].count())

#calculate total purchase value
total_paid_sn = sn_grouped["Price"].sum()

#calculate average purchase price
avg_price_sn = round(sn_grouped["Price"].mean(),2)

#add columns to the SN grouped DataFrame
sn_grouped_df["Average Purchase Price"] = avg_price_sn.map("${:,.2f}".format)
sn_grouped_df["Total Purchase Value"] = total_paid_sn.map("${:,.2f}".format)

#rename column Purchase ID to Purchase Count
renamed_sn_grouped_df = sn_grouped_df.rename(columns={"Purchase ID":"Purchase Count"})
renamed_sn_grouped_df

#create sorted DataFrame by Purchase Count
top_spenders_count = renamed_sn_grouped_df.sort_values("Purchase Count", ascending=False)

#create sorted DataFrame by Total Purchase Value based on sorted count values
top_spenders_value = top_spenders_count.sort_values("Total Purchase Value", ascending=False)
top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create group by object for Item ID and Item Name
popular_items_grouped_df = purchase_data.groupby(["Item ID","Item Name"])

#create a new data frame for the summary
top_items_df = pd.DataFrame(popular_items_grouped_df["Purchase ID"].count())
top_items_df

#calculate total purchase value
total_paid_item = popular_items_grouped_df["Price"].sum()

#calculate Item Price
item_price = popular_items_grouped_df["Price"].mean()

#add columns to top_items_df
top_items_df["Item Price"] = item_price.map("${:,.2f}".format)
top_items_df["Total Purchase Value"] = total_paid_item.map("${:,.2f}".format)

#rename purchase ID to Purchase Count
renamed_top_items_df = top_items_df.rename(columns={"Purchase ID":"Purchase Count"})
renamed_top_items_df

#organize by purchase count
top_items_df_2 = renamed_top_items_df.sort_values("Purchase Count", ascending=False)
top_items_df_2.head()
```
